```python
#Encoding
#Coverting all data types of input columns to numerical columns
#Here all the columns are numerical. So no need of encoding.

```

```python
#Separating input and output columns
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#dimension of x
x.ndim
```

```python
#dimension of y
y.ndim
```

```python
#Converting to training and testing datas
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Normalization
#Converting all the input columns to same scale
#Here glucose column has high range values and diabeties predegree function has lower values.
#So we need to convert all the columns to same scale.
#2 methods,
    #1.Standard Scaler
    #2.Min-Max Scaler
```

```python
#Standard Scaler
#z=(x-u)/s
#z=output
#x=input
#u=mean of x(training data)
#s=standard deviation of x(training data)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#Model creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
#To give random input values
print(knn.predict(scaler.transform([[3,148,80,20,120,40.9,0.4568,47]])))

```

```python
#Model performance Evaluation
#------------------------------
#1. Accuracy Score
#2. Recall
#3. Precision
#4. F1 Score

#Based on Confusion Matrix all these performance metrics are calculated.

#It is used to evaluate the performance of classification model.
```

```python
#4 technical terms in confusion matrix
#1. TP : True Positive  ==> Prediction and actual are true ==> prediction 1 and actual 1
#2. TN : True Negative  ==> Prediction and actual are false ==> prediction 0 and actual 0
#3. FP : False Positive ==> Prediction is true but actual is false ==> prediction 1 and actual 0
#4. FN : False Negative ==> Prediction is false but actual is true ==> prediction 0 and actual 1
```

```python
#Shape of the confusion matrix is the number of predictions
#For a diabetic patient, there are 2 possibilities. So the shape of confusion matrix is 2*2
#majority of the times confusion matrix is 2*2
#In a confusion matrix,
    # TN    FP
    # FN    TP
```

```python
#y_test : 1 0 1 0 1 1 1 1 0 0 0 1 0 1 0 1 1 0
#y_pred : 1 0 1 1 1 0 1 1 0 1 1 0 1 1 0 1 1 0

#TP=8
#TN=4
#FP=4
#FN=2

#[4 4]
#[2 8]
```

```python
#Accuracy Score
#It is the ratio of correctly predicted observation to the total observations.
#score=(TP+TN)/(TP+TN+FP+FN)
#If  the acuracy score is above 70% means the model is good.
#Otherwise the model has some drawbacks. We have to check it again.
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
#Each time the accuracy score may vary because the train_test_split function randomly splits the data into training and testing data.
#Some times it takes first 70% data for training and remaining 30% for testing.
#Or it may take last 70% data for training and remaining 30% for testing.
#Or it may take any random 70% data for training and remaining 30% for testing.
#So the accuracy score may vary each time.
```

```python
#To get consistent accuracy score, we use random_state parameter in train_test_split function.
#Experimentally it is found that random_state=42, 0 or 1 gives good accuracy score. 
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.naive_bayes import GaussianNB
base=GaussianNB()
from sklearn.svm import SVC
sv=SVC()
lst=[knn,base,sv]
```

```python
for i in lst:
    print("Model name is",i)
    i.fit(x_train,y_train)
    y_pred=i.predict(x_test)
    y_pred
    print(confusion_matrix(y_test,y_pred))
    print(accuracy_score(y_test,y_pred))
    print(classification_report(y_test,y_pred))
```
